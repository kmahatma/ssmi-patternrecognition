```python
import cv2
import matplotlib.pyplot as plt

# Define directory path
file_path = "/content/drive/MyDrive/@-ssmi-pattern-recognition/prak-02/"

# Read the image
babon = cv2.imread(file_path + "babon.jpeg")

# Check if the image is loaded correctly
if babon is None:
    print("Error: Image not found at", file_path + "babon.jpg")
else:
    # Convert to grayscale
    babon_gray = cv2.cvtColor(babon, cv2.COLOR_BGR2GRAY)

    # Convert BGR to RGB for correct Matplotlib display
    babon_rgb = cv2.cvtColor(babon, cv2.COLOR_BGR2RGB)

    # Display images side by side
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(babon_rgb)  # Use RGB image
    plt.title("Original Image")
    plt.axis("off")  # Hide axis

    plt.subplot(1, 2, 2)
    plt.imshow(babon_gray, cmap="gray")  # Use grayscale colormap
    plt.title("Grayscale Image")
    plt.axis("off")  # Hide axis

    plt.show()

    # Access pixel values at a specific position (e.g., (100,100))
    pixel_value = babon[100, 100]  # BGR values
    print(f"Pixel value at (100,100) [BGR]: {pixel_value}")
```
