**2.3.B Histogram**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data_path = "/content/drive/MyDrive/Colab Notebooks/ssmi-pattern-recognition/prak-03-data/"  # Change this if needed
data = pd.read_csv(data_path + "tips.csv")

# Set the style for better visualization
sns.set_style("whitegrid")

# Create histogram with seaborn
plt.figure(figsize=(8, 5))
sns.histplot(data, x="tip", hue="sex", kde=True, bins=15, multiple="stack", palette="pastel")

# Set labels and title
plt.xlabel("Tip Amount ($)")
plt.ylabel("Frequency")
plt.title("Histogram of Tips by Gender")

# Show the plot
plt.show()
```
